```python
import torch
from torch.optim import LBFGS

def objective_function(w):
  return w[0]*torch.exp(w[0]) + w[1]*torch.exp(w[1])


w = torch.tensor([0.5, -0.5], requires_grad=True)
optimizer = LBFGS([w], lr=0.1, max_iter=100, tolerance_grad=1e-5, tolerance_change=1e-5)

def closure():
  optimizer.zero_grad()
  loss = objective_function(w)
  loss.backward()
  return loss

# optimize
optimizer.step(closure)

print("Optimized w : ", w.detach())
print("Minimm function value:", objective_function(w).item())
```
