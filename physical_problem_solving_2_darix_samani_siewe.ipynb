```python
import numpy as np
import matplotlib.pyplot as plt

@np.vectorize
def solution(t):
    """
    This function is the analytical soltuion of uour first differentiel equation
    """
    return np.cos(t)


def f(x):
    """
    the function f(t,x,u) its depends only on x in our case
    """
    w_0 = 1
    return - (w_0**2)*x
h = 0.01

time = np.arange(0, 20, h)

x, u = np.zeros_like(time), np.zeros_like(time)


x[0] = 1.0
u[0] = 0.0

for i, _ in enumerate(time[:-1]):
    x[i+1] = x[i] + h*u[i]
    u[i+1] = u[i] + h*f(x[i])

plt.figure(figsize=(15,8))
plt.plot(time, x)
plt.plot(time,solution(time))
plt.ylabel('displacement x')
plt.xlabel('time t')
plt.legend(["Euler method with $h=0.01$", "exact solutions"])
plt.grid()
plt.savefig("methode_euler_h_0.01.png")
plt.show()
    
```

```python

import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt 
x_0 = [1, 0]

def diff(y, t, w_0=1):
    
    x, u = y
    dxdt = [u, -(w_0**2)*x]
    return dxdt
    
t = np.linspace(0, 10, 101)

plt.figure(figsize=(15,8))
sol = odeint(diff, x_0, t, args=(1,),)
plt.plot(t, sol[:, 0], 'b', label='x(t)')
plt.plot(t, sol[:, 1], 'g', label='v(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.savefig("odeint_2a.png")
plt.show()
```

```python
@np.vectorize
def U(x, w_0=1):
    return 0.5*(w_0**2)*x**2

@np.vectorize
def K(v, m=1):
    return 0.5*m*v**2


plt.figure(figsize=(15,8))
plt.plot(t, U(sol[:, 0]), 'b', label='U(t)')
plt.plot(t, K(sol[:, 1]), 'g', label='K(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.savefig("odeint_2b.png")
plt.show()
```
