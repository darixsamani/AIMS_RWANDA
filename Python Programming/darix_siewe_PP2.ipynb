2. Write a function that given two integers b and n (n is positive), calculates bn without using
python’s power operator. Name this function ownpowerfn.py.
PS: What if we do not specify that n has to be positive?

```python
from math import prod
def ownpowerfn():
    """
    The propulse of this function is to calculate the power of b^n without using python power operator
    
    Input user: we ask to use to enter value of b and n (integer)
    
    Output: it is important to note that our function handle the case when n is negative
    
    """
    
    b = int(input("Enter the integer b: "))
    n = int(input("Enter the integer n: "))
    power = prod([b for _ in range(n if n>=0 else -n)])
    return power if n>=0 else 1/power

print("Try of n positive ", ownpowerfn())
print("Try of n negative ", ownpowerfn())
```

```python
def arithmetic_mean(a: float, b: float)-> float:
    """
    the aims of this function is to calculate the arithmetic means given two number
    
    Agrs():
        a (float): the first value of our function
        b (float): the second value of our function
    
    Retruns:
        float: the arithmetic means given two numbers
        
    """
    return (a+b)/2

def geometric_mean(a: float, b: float)-> float:
    
    """
    the aims of this function is to calculate the geometric means given two number
    
    Agrs():
        a (float): the first value of our function
        b (float): the second value of our function
    
    Retruns:
        float: the geometric means given two numbers
        
    """
    
    return (a*b)**(0.5)
    
print(arithmetic_mean(1, 2))
print(geometric_mean(1, 2))
```

```python
from math import sqrt
def distancefn(a, b)-> float:
    """
    The aims of this function is to calcluate the distance between two poitns given their coordinates
    
    Agrs:
        a (tuple or list of two numerical value): the coordinate of the first point in this space
        a (tuple or list of two numerical value): the coordinate of the second point in this space
    
    Returns:
        (float): the distance of the point a and b
    """
    return sqrt((a[0]+b[0])**2 + (a[1]+b[1])**2)

def perimeter_of_triangle(a, b, c)-> float: 
    """
    The propulse of this function is to calculate the perimter of triangle given the coordinates of this three points
    to achieve this, we use the previous function the calculate the distance between two point and after sum the three sides
    of our triangle
    
    Args:
        a (tuple or list of two numerical value): the coordinate of the first point of triangle
        b (tuple or list of two numerical value): the coordinate of the second point of traingle
        c (tuple or list of two numerical value): the coordinate of the first point of triangle
        
    Returns 
        (float): perimeter of the triangle
        
        
        
        
        
    """
    return distancefn(a, b) + distancefn(a, c)+ distancefn(b,c)

def area_of_triangle(a, b, c)-> float:
    """
       The propulse of this function is to calculate the area of the traingle gievn the coordinates in the space of his three point
       
       Args:
        a (tuple or list of two numerical value): the coordinate of the first point of triangle
        b (tuple or list of two numerical value): the coordinate of the second point of traingle
        c (tuple or list of two numerical value): the coordinate of the first point of triangle
        
        Returns:
            (float): the area of the traingle
           
    """
    return abs(0.5*(a[0]*(b[1] - c[1]) + b[0]*(c[1]-a[1]) + c[0]*(a[1]-b[1])))

print(distancefn((1,2), (2,4)), perimeter_of_triangle((1,2), (4,-1), (1,1)), area_of_triangle((1,2), (4,-1), (1,1)))
```

6. For a fixed real number x and a natural number n, we can define recursively xn using the
relations:

$x_0 = 1$ and $x_{n+1} = x*x_n$.
Write a function power(x,n) that implements the above recursion.

```python
def power(x: float,n: int):
    """
    The propulse of this function is to use recursion in order to calculte the power x^n
    it is important to note that this function don't handle the case when n is negative
    
    Agrs:
        x (float): value of x
        n: (int): the value of n(positive)
    
    Returns:
        float: the value of x^n
    """
    return 1 if n == 0 else x*power(x, n-1)
print("The alue of 3^2 is : ", power(3,2))
```

alphabetical order merges them into a single string in increasing alphabetical order.
For example `M(‘bd′, ‘abcf′) = ‘abbcdf′`

```python
from string import ascii_letters
def M(s_1: str, s_2: str)-> str:
    """
    
    """
    s = ""
    for l in ascii_letters:
        if l in s_1:
            s+=l*s_1.count(l)
        if l in s_2:
            s+=l*s_2.count(l)
    return s
print(M("bd", "abcf"))
```
