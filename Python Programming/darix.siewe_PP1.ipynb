## About Random number generator

Investigate how python generates random numbers.

python have module random to help us to generate random number. Let's explore this module with some example

````python
import random

# if we want to generate a random number(floating point) from 0 to 1 we can use random.random() function
# example

n = random.random()
print("A random number generate from 0 to 1: ", n)

## we can also generate a natural number between two integer using randrange

n = random.randrange(0, 100)

print("A random number from 0 to 100", n)

## we can use a thrid paramter of randrange to specifie the divisibility of number generate
## example to generate even number

n = random.randrange(0, 100, 2)

print("A random even number from 0 to 100", n)

## example to geneate number such that it is multiple of 3
n = random.randrange(0, 100, 3)

print("A random number from 0 to 100 that is multiple of 3: ", n)


````
