**Exercise 3**

Consider the system

\begin{equation}
    \begin{cases}
        2x_1+7x_2=3-x_3\\
        -x_1+2x_3=8-6x_2\\ \label{eq1} \tag{1}
        x_2+8x_3=0\\
        5x_1+9x_2+10x_3 = -1\\
        4x_1+3x_2=5-5x_3
    \end{cases}
\end{equation}
write it in the form of $Ax=b$.


Use **QR factorization** to solve the system \eqref{eq1} 


```julia
A = [2 7 1; -1 6 2; 0 1 8; 5 9 10; 4 3 5]
b = [3; 8; 00; -1; 5]

QR = qr(A)
R = Matrix(QR.R)
Q = Matrix(collect(QR.Q))
y =  Q'*b
x = R\y[1:3]
println("The solution using qr decomposition is : ", x)
```
